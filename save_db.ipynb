```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import sqlalchemy as sqla
```

```python
con=sqla.create_engine('sqlite:///example_data//django_bus_sys')
```

```python
df = pd.DataFrame({'id': [1, 2, 3, 4], 'name': ['zhangsan', 'lisi', 'wangwu', 'zhuliu']})
```

```python
df
```

```python
df.to_sql('test_person', con, if_exists='replace', index=True)
```

```python
xls=pd.ExcelFile("example_data//一公司安全员工信息.xls")
item=[]
for sheet in xls.sheet_names:
    item.append(pd.read_excel(xls,"一车队"))
result=pd.concat(item)
```

```python
result.columns=list(range(len(result.columns)))
result.index=range(len(result))
cleaned=result.rename(columns={0:'person_id',1:'name',2:'job',3:'celphone',4:'time_in_company',5:'belong_route'})
```

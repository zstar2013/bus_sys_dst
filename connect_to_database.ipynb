```python
import sqlalchemy as sqla
import pandas as pd
```

```python
db = sqla.create_engine('sqlite:///example_data//django_bus_sys')
```

```python
car_info=pd.read_sql('select sub_car_id,cartype_id,car_id from bus_info_businfo', db)
```

```python
type_info=pd.read_sql('select id,target_value1,target_value2 from bus_info_cartype', db)
```

```python
pd.merge(car_info,type_info,left_on="cartype_id",right_on="id",how='left')
```
